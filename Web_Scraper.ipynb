```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd

url = ""
filename = ""

# print(columns)

# create/list columns for data
column_names = ["Vehicle",
                "Blank1",
                "All_MPG",
                "Blank2",
                "Combined_MPG",
                "MPG_Type",
                "City_MPG",
                "HWY_MPG",
                "Blank3",
                "Text1",
                "Text_city",
                "Text_hwy",
                "Fuel_Rate",
                "Text2",
                "Blank4",
                "Annual_Fuel_Cost",
                "MSRP",
                "Car_Spec1",
                "Car_Spec2",
                "Total_Range"
]

# create dictionary for data to load into:
car_dict = {}

def web_scraper(url,filename):
    # load request data to variable and parse data via BeautifulSoup
    response = requests.get(url)
    results = soup(response.content, "html.parser")

    # filter table data "td" and clean data
    table_data = results.find_all("td")
    columns= [value.text.replace("\n","").replace("\t","").replace("\r","") for value in table_data]


    # enumerate through column_names to load dictionary with the correct and value from columns data
    # since data is in a list, enumerate through and skip every 20 because 20 columns,
    # load into dataframe to load into SQL
    for idx, key in enumerate(column_names):
        car_dict[key] = columns[idx:][::20]

        df = pd.DataFrame(car_dict)
        # print(df)

    df.to_csv(filename,index=False)
```

```python
# insert data into SQLite

import sqlite3
import pandas as pd

df = pd.read_csv("Electric_Hybrid_Gas_Cars_Combined.csv")
connection = sqlite3.connect("Capstone.db")
cursor = connection.cursor()

# Clean MSRP column. Take highest value if there is an MSRP range
for idx in df.index:
    if "–" in df.MSRP[idx]:
        MSRP_Updated = df.MSRP[idx].split("–")
        MSRP_Updated = MSRP_Updated[1]
        df.MSRP[idx] = MSRP_Updated[1:]
    else:
        df.MSRP[idx] = df.MSRP[idx]

# Removing text from Total_Range column
for str in df.index:
    if "milesTotal" in df.Total_Range[str]:
        T_Range_Updated = df.Total_Range[str].split(" milesTotal Range")
        output = ""
        for value in T_Range_Updated:
            output += value
        df.Total_Range[str] = output[1:]
    else:
        df.Total_Range[str] = df.Total_Range[str]
# Clean TotalRange column, take highest value if there is a range of values
for idx in df.index:
    if "-" in df.Total_Range[idx]:
        Total_Range1 = df.Total_Range[idx].split("-")
        Total_Range1 = Total_Range1[1]
        df.Total_Range[idx] = Total_Range1[1:]
    else:
        df.Total_Range[idx] = df.Total_Range[idx]

# If range was not provided in Total_Range, replace with 0
for idx in df.index:
    if "Range not available" in df.Total_Range[idx]:
        df.Total_Range[idx] = 0
    else:
        df.Total_Range[idx] = df.Total_Range[idx]

# Initial drop of columns that aren't needed
df = df.drop(['Blank1','Blank2','Blank3','Text1','Text_city','Text_hwy','Blank4',],axis=1)

df.to_sql('Vehicle_Data',connection,if_exists='replace')


connection.close()
```

```python
# iterate through rows of data to restructure how cost is shown
import pandas as pd

import pandas as pd
import sqlite3

connection = sqlite3.connect("Capstone.db")

dfDB = pd.read_sql_query("select * from Vehicle_Data",connection)

Vehicle = []
PaymentYear = []
TotalPayment = []

VehicleCost = pd.DataFrame(list(zip(Vehicle, PaymentYear,TotalPayment)),
               columns =['Vehicle', 'PaymentYear','TotalPayment'])

for idx in dfDB.index:
    # print(dfDB.Vehicle[idx])
    # print(dfDB.MSRP[idx])
    # print(dfDB.MSRP[idx]+dfDB.Annual_Fuel_Cost[idx])
    i = 6
    x = 1
    while x < 6:
        AppendList = [
                dfDB.Vehicle[idx],
                x,
                (dfDB.MSRP[idx]/5)*x+dfDB.Annual_Fuel_Cost[idx]*x,
                ]
        x=x+1
        VehicleCost.loc[len(VehicleCost)] = AppendList
    while i < 13:
        AppendList = [
                dfDB.Vehicle[idx],
                i,
                dfDB.MSRP[idx]+dfDB.Annual_Fuel_Cost[idx]*i,
                ]
        i=i+1
        VehicleCost.loc[len(VehicleCost)] = AppendList

VehicleCost.to_csv("VehicleCostData_Backup.csv",index=False)
VehicleCost.to_sql('Vehicle_Cost',connection,if_exists='replace')


```
